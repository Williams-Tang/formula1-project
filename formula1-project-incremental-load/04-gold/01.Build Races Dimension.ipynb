```python
from pyspark.sql import functions as F
```

```python
target_table = f'{catalog_name}.{gold_schema}.dim_races'
```

**Step 1 - Read source tables**
- `circuits`
- `races`

```python
circuits_df = (
    spark.read.table(f"{catalog_name}.{silver_schema}.circuits")
              .filter(F.col("batch_id") == v_batch_id)
)
```

```python
races_df = (
    spark.read.table(f"{catalog_name}.{silver_schema}.races")
         .filter(F.col("batch_id") == v_batch_id)
)
```

**Step 2 - Join `races` with `circuits` using `circuit_id`**

```python
dim_races_df = (
            races_df
                .join(circuits_df, 
                    circuits_df.circuit_id == races_df.circuit_id,
                    "inner")
                .select(
                    races_df.season,
                    races_df.round,
                    races_df.race_name,
                    races_df.race_date,
                    circuits_df.circuit_name,
                    circuits_df.locality,
                    circuits_df.country
                )
)
```

```python
display(dim_races_df)
```

**Step 3 - Write the transformed data to the `gold` `dim_races` table**

```python
write_to_gold(
    input_df = dim_races_df,
    target_table = target_table,
    merge_condition = "t.season = s.season AND t.round = s.round",
    columns_to_update = [
        "race_name",
        "race_date",
        "circuit_name",
        "locality",
        "country"
    ]
)
```
