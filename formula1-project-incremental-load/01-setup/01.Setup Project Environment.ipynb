#Set up the project environment for the project
1. Create External Location databricks-course-ext-dl628-formula1-incr

**Access Cloud Storage**

```sql
%fs ls 'abfss://formula1-incr@databrickssourceextdl628.dfs.core.windows.net/'
```

**Create External Location**

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS databricks_course_ext_dl628_formula1_incr
URL 'abfss://formula1-incr@databrickssourceextdl628.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL `databricks-course-sc`)
COMMENT 'External location for the formula1 container';
```

**Create Catalog Formula1**

```sql
SHOW CATALOGS
```

```sql
CREATE CATALOG IF NOT EXISTS formula1_incr
   MANAGED LOCATION 'abfss://formula1-incr@databrickssourceextdl628.dfs.core.windows.net/'
   COMMENT 'This is the main catalog for the formula1 project' ;
```

**Create Schemas landing, bronze, silver, gold**

```sql
CREATE SCHEMA IF NOT EXISTS formula1_incr.landing;
CREATE SCHEMA IF NOT EXISTS formula1_incr.bronze
   MANAGED LOCATION 'abfss://formula1-incr@databrickssourceextdl628.dfs.core.windows.net/bronze';
CREATE SCHEMA IF NOT EXISTS formula1_incr.silver
   MANAGED LOCATION 'abfss://formula1-incr@databrickssourceextdl628.dfs.core.windows.net/silver';
CREATE SCHEMA IF NOT EXISTS formula1_incr.gold
   MANAGED LOCATION 'abfss://formula1-incr@databrickssourceextdl628.dfs.core.windows.net/gold';
```

**Create Volume Files**

```sql
CREATE EXTERNAL VOLUME formula1_incr.landing.files
LOCATION 'abfss://formula1-incr@databrickssourceextdl628.dfs.core.windows.net/landing';
```
